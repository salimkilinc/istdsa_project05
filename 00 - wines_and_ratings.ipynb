# Preprocessing

## Replacing Some of Special Characters in Specified Columns

```python
wines_and_ratings["Body"] = wines_and_ratings["Body"].str.replace("-", " ")
wines_and_ratings["Body"].head(5)
```

```python
wines_and_ratings["Grapes"] = wines_and_ratings["Grapes"].str.replace("[", " ").str.replace("]", " ").str.replace("'", "").str.strip()
wines_and_ratings["Grapes"].head(5)
```

```python
wines_and_ratings["Harmonize"] = wines_and_ratings["Harmonize"].str.replace("[", " ").str.replace("]", " ").str.replace("'", "").str.strip()
wines_and_ratings["Harmonize"].head(5)
```

```python
wines_and_ratings
```

## Splitting Specified Columns by a Delimiter and Expanding Them into Multiple Columns

```python
def split_and_expand_columns(df, column_name, delimiter="/"):
    expanded_columns = df[column_name].str.split(delimiter, expand=True)
    new_columns = [f"{column_name}{i+1}" for i in range(expanded_columns.shape[1])]
    expanded_columns.columns = new_columns

    df = pd.concat([df, expanded_columns], axis=1)
    df = df.fillna("")
    df = df.drop(columns=[column_name])

    return df
```

```python
wines_and_ratings = split_and_expand_columns(wines_and_ratings, "Type")
```

```python
wines_and_ratings = split_and_expand_columns(wines_and_ratings, "Elaborate")
```

```python
wines_and_ratings = split_and_expand_columns(wines_and_ratings, "Grapes", delimiter=", ")
```

```python
wines_and_ratings = split_and_expand_columns(wines_and_ratings, "Harmonize", delimiter=", ")
```

```python
wines_and_ratings.info()
```

## Transforming Unreasonable Values in the ABV Column

```python
def transform_values(df, column_name):
    df[column_name] = np.where(df[column_name] > 14.5, 14.5, df[column_name])
    df[column_name] = np.where(df[column_name] < 9, 9, df[column_name])
    return df
```

```python
wines_and_ratings = transform_values(wines_and_ratings, 'ABV')
wines_and_ratings['ABV'].sort_values()
```

## Checking and Replacing Special Characters in Some Other Columns

```python
def check_special_characters(df, column_name, special_characters_pattern):
    rows_with_special_characters = df[df[column_name].str.contains(special_characters_pattern, regex=True, na=False)]
    row_count = len(rows_with_special_characters)
    print("Number of rows with special characters in {}: {}".format(column_name, row_count))
    print(rows_with_special_characters[[column_name]].drop_duplicates().sort_values(by=column_name))
```

```python
import re
def remove_special_characters(df, column_name, characters_to_remove):
    df[column_name] = df[column_name].str.replace(f'[{re.escape(characters_to_remove)}]', ' ', regex=True)
    return df
```
